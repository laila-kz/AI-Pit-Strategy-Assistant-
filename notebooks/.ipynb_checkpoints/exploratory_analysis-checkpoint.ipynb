```python
#clean up provisional df 
# 1️⃣ Normalize column names
provisional_results.columns = provisional_results.columns.str.strip().str.upper()

# 2️⃣ Convert FL_TIME and TOTAL_TIME to seconds
def time_to_seconds(t):
    """
    Convert 'M:SS.mmm' or 'MM:SS.mmm' to seconds (float).
    Handles None or '-' gracefully.
    """
    if pd.isna(t) or t in ['-', '']:
        return None
    try:
        parts = str(t).split(':')
        if len(parts) == 2:
            m, s = parts
            return float(m) * 60 + float(s)
        return float(t)
    except:
        return None

provisional_results['FL_TIME_SEC'] = provisional_results['FL_TIME'].apply(time_to_seconds)
provisional_results['TOTAL_TIME_SEC'] = provisional_results['TOTAL_TIME'].apply(time_to_seconds)

# 3️⃣ Convert GAP_FIRST to numeric (e.g. "+0.652" → 0.652 sec)
def gap_to_seconds(gap):
    if pd.isna(gap) or gap in ['-', '']:
        return 0.0  # Leader
    try:
        return float(gap.replace('+', ''))
    except:
        return None

provisional_results['GAP_FIRST_SEC'] = provisional_results['GAP_FIRST'].apply(gap_to_seconds)

# 4️⃣ Drop redundant constant columns (optional)
provisional_results = provisional_results.drop(columns=['STATUS', 'CLASS', 'DIVISION', 'VEHICLE'])

# 5️⃣ Verify results
provisional_results.info()
display(provisional_results.head())
```

#### Detect Outliers:
Outliers are data points that are significantly different from other observations in a dataset, appearing as unusually high or low values


```python
#detect outliers 
#The goal of this code is to identify which individual lap times are significantly faster or slower than the typical laps, using a robust statistical definition of "outlier".
Q1= lap_duration['lap_time_sec'].quantile(0.25)
Q3=lap_duration['lap_time_sec'].quantile(0.75)
IQR = Q3 - Q1

#define outliers 
outliers = lap_duration[
(lap_duration['lap_time_sec'] < Q1 - 1.5 * IQR) |
    (lap_duration['lap_time_sec'] > Q3 + 1.5 * IQR)
]

outliers[['vehicle_id', 'lap', 'lap_time_sec']]
```

```python
#Make Sure Data Is Clean
import pandas as pd
import numpy as np

# Normalize column names
# This step is good practice and ensures all columns are uppercase and stripped of whitespace
provisional_results.columns = provisional_results.columns.str.strip().str.upper()

# Convert time fields to seconds
def time_to_seconds(t):
    """
    Converts time strings (MM:SS.mmm or HH:MM:SS.mmm) to total seconds (float).
    Handles empty strings or non-standard formats gracefully by returning None.
    """
    if pd.isna(t) or t == '':
        return None
    try:
        parts = str(t).split(':')
        if len(parts) == 3:
            # HH:MM:SS.mmm format
            return int(parts[0]) * 3600 + int(parts[1]) * 60 + float(parts[2])
        elif len(parts) == 2:
            # MM:SS.mmm format
            return int(parts[0]) * 60 + float(parts[1])
        else:
            # Already a numeric string representing seconds
            return float(t)
    except:
        # Return None if conversion fails (e.g., status like 'Retired')
        return None

provisional_results['FL_TIME_SEC'] = provisional_results['FL_TIME'].apply(time_to_seconds)
provisional_results['TOTAL_TIME_SEC'] = provisional_results['TOTAL_TIME'].apply(time_to_seconds)

# Convert numeric columns
provisional_results['POSITION'] = pd.to_numeric(provisional_results['POSITION'], errors='coerce')

# --- FIX APPLIED HERE ---
# Changed 'TOTAL_LAPS' to the correct column name 'LAPS'
provisional_results['LAPS'] = pd.to_numeric(provisional_results['LAPS'], errors='coerce')
```

#####  Derive features inside lap data

```python
# Convert lap time value (if in ms) to seconds
lap_time['lap_time_s'] = lap_time['value'] / 1000.0

# Sort for lag operations
lap_time = lap_time.sort_values(['vehicle_id', 'timestamp'])

# Calculate time difference between consecutive laps per vehicle
lap_time['lap_interval_s'] = lap_time.groupby('vehicle_id')['timestamp'].diff().dt.total_seconds()

# Create consistency check flag
lap_time['is_outlier'] = (lap_time['lap_time_s'] < 1) | (lap_time['lap_time_s'] > lap_time['lap_time_s'].quantile(0.99))

print("✅ Derived features added to lap_time: ['lap_time_s', 'lap_interval_s', 'is_outlier']")
```

##### Merge with nearest weather info (within 60s window)

```python
def get_nearest_weather(ts):
    diffs = (weather_race['TIME_UTC_STR'] - ts).abs()
    idx = diffs.idxmin()
    if diffs.min().total_seconds() <= 60:  # only if within 1 minute
        return weather_race.loc[idx, ['AIR_TEMP', 'HUMIDITY', 'WIND_SPEED', 'RAIN']]
    else:
        return pd.Series([np.nan, np.nan, np.nan, np.nan], index=['AIR_TEMP', 'HUMIDITY', 'WIND_SPEED', 'RAIN'])

lap_time[['AIR_TEMP', 'HUMIDITY', 'WIND_SPEED', 'RAIN']] = lap_time['timestamp'].apply(get_nearest_weather)
print("✅ Weather columns merged into lap_time.")
```

##### Merge with provisional results (by vehicle ID or car number)

```python
# Normalize vehicle_id if possible (extract numeric car number)
lap_time['car_number'] = lap_time['vehicle_id'].str.extract(r'(\d+)$').astype(float)

# Match with provisional
merged = pd.merge(
    lap_time,
    provisional_results,
    left_on='car_number',
    right_on='NUMBER',
    how='left',
    suffixes=('', '_prov')
)

print(f"✅ Merged with provisional results → {merged.shape[0]} rows, {merged.shape[1]} columns")
```

##### Derived performance features

```python
# ✅ 4️⃣ Add derived performance features
merged['lap_rank'] = merged.groupby('vehicle_id')['lap_time_s'].rank()
merged['is_fastest_lap'] = merged['lap_rank'] == 1
merged['avg_speed_est'] = 4.0 / (merged['lap_time_s'] / 3600)  # Example: 4 km/lap
print("✅ Derived performance features added.")

# 🧹 5️⃣ Final dataset cleanup
final_cols = [
    'vehicle_id', 'lap', 'lap_time_s', 'lap_interval_s', 'is_outlier',
    'AIR_TEMP', 'HUMIDITY', 'WIND_SPEED', 'RAIN',
    'POSITION', 'NUMBER', 'LAPS', 'TOTAL_TIME', 'FL_TIME',
    'avg_speed_est', 'is_fastest_lap', 'FL_KPH', 'FL_TIME_SEC',
    'TOTAL_TIME_SEC', 'GAP_FIRST_SEC'
]

# Keep only the columns that exist in our merged DataFrame
final_cols = [c for c in final_cols if c in merged.columns]

final_dataset = merged[final_cols]

print(f"✅ Final dataset ready with {final_dataset.shape[0]} rows and {final_dataset.shape[1]} columns.")
```
